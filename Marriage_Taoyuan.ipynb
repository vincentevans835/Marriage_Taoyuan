```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

table = pd.read_csv('marriageTaoyuanData.csv')

plt.figure(figsize = (16,16)) 

plt.subplot(2,2,1)

width = 0.3

plt.barh(table['Year'], table['Total - Male'],width,label='Male')

plt.barh(table['Year'] + width, table['Total - Female'],width,label='Female')

plt.title('Total Marriage Population')

plt.ylabel('Year')

plt.xlabel('Population')

plt.legend()

plt.subplot(2,2,2)

width = 0.3

plt.bar(table['Year'], table['Foreign Nationality - Male'],label='Foreign Nationality Male')

plt.bar(table['Year'], table['Mainland Hong Kong and Macao Regions - Male'],label='Mainland Hong Kong and Macao Regions Male')

plt.ylabel('Year')

plt.xlabel('Population')

plt.legend()

plt.savefig('410821323_mid-1Matplotlib.jpg')

table
```
